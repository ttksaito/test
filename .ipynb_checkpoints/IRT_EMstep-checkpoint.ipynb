## 関数の定義

```python
def rounding(x):
    return Decimal(str(x)).quantize(Decimal('0.0001'), rounding=ROUND_HALF_UP)

# 　　　項目情報関数（IIC）
# 　　　theta　：　特性値　　　p　：　項目母数の個数　　　 j　：　j番目の項目
def P(theta,a,b,c):
    return c+(1-c)*(1/(1+math.exp(-1.7*a*(theta-b))))

def Q(theta,a,b,c):
    return 1-P(theta,a,b,c)

#  特性値がθの受験者が項目パタンuを取る確率（尤度関数）
#  u　：　項目反応パタン [1,0,1,0]
# theta　：　特性値 θ
# abc_para　：　母数パラメタ [ [a,b,c],[a,b,c],[a,b,c],[a,b,c] ]
def L(u,theta, abc_para):

    dot_list=[]
    for j in range(len(abc_para)):
        a,b,c=abc_para[j]
        dot_list.append(P(theta,a,b,c)**u[j]*((1-P(theta,a,b,c))**(1-u[j])))
    
    return np.prod(dot_list)

def f(x):
    return 1/math.sqrt(math.pi)*math.exp(-x**2/2)
```

## G_imを返す関数の定義

```python
def G_im(abc_para):
    
    u_list=[df.iloc[i,0:len(q_num_list)].values.tolist() for i in range(examinee_num)]

    c=sum([f(X_m) for X_m in x_plot_list])

    G_im_list=[]
    for i in range(examinee_num):
        u=u_list[i]
        denominator_i=sum([L(u,X_m, abc_para)*f(X_m)/c for X_m in x_plot_list])
        G_im_list.append([L(u,X_m, abc_para)*f(X_m)/c/denominator_i for X_m in x_plot_list])

    #　200人分のG_imからNとrを求めるので、先にG_imを算出する
    return G_im_list#G_im_listは200行15列
```

## 勾配ベクトルの定義

```python
def A(X_m, b):
    return X_m-b

def B(X_m, abc_para):
    a,b,c=abc_para
    return P(X_m,a,b,c)-c

def C(r,N,X_m, abc_para):
    a,b,c=abc_para
    return r-N*P(X_m,a,b,c)

def matrix_d(abc_para,r_list):

    a,b,c=abc_para
    
    element1=1.7/(1-c)*sum([A(x_plot_list[m], b)*B(x_plot_list[m], abc_para)*C(r_list[m],N_list[m],x_plot_list[m], abc_para)/P(x_plot_list[m],a,b,c) for m in range(div)])
    element2=-1.7*a/(1-c)*sum([B(x_plot_list[m], abc_para)*C(r_list[m],N_list[m],x_plot_list[m], abc_para)/P(x_plot_list[m],a,b,c) for m in range(div)])
    # 1/(1-c)*sum([C(r_list[m],N_list[m],x_plot_list[m], abc_para)/P(x_plot_list[m],a,b,c) for m in range(div)])
    
    return np.matrix([[element1],[element2]])
```

## 情報関数の定義

```python
def rounded_a(X_m,abc_para):
    a,b,c=abc_para
    return 1.7*A(X_m,b)*B(X_m, abc_para)*Q(X_m,a,b,c)/(1-c)

def rounded_b(X_m,abc_para):
    a,b,c=abc_para
    return -1.7*a*B(X_m, abc_para)*Q(X_m,a,b,c)/(1-c)

def rounded_c(X_m,abc_para):
    a,b,c=abc_para
    return (1-P(X_m,a,b,c))/(1-c)

def matrix_I(abc_para):
    
    a,b,c=abc_para
    
    element11=sum([N_list[m]/(P(x_plot_list[m],a,b,c)*Q(x_plot_list[m],a,b,c))*rounded_a(x_plot_list[m],abc_para)**2 for m in range(div)])
    element12=sum([N_list[m]/(P(x_plot_list[m],a,b,c)*Q(x_plot_list[m],a,b,c))*rounded_a(x_plot_list[m],abc_para)*rounded_b(x_plot_list[m],abc_para) for m in range(div)])
    element21=sum([N_list[m]/(P(x_plot_list[m],a,b,c)*Q(x_plot_list[m],a,b,c))*rounded_b(x_plot_list[m],abc_para)*rounded_a(x_plot_list[m],abc_para) for m in range(div)])
    element22=sum([N_list[m]/(P(x_plot_list[m],a,b,c)*Q(x_plot_list[m],a,b,c))*rounded_b(x_plot_list[m],abc_para)**2 for m in range(div)])

    return np.matrix([[element11,element12],[element21,element22]])
```

## 通過率と点双列相関係数のdfを返す関数の定義

```python
def PR_r(df):
    #通過率
    pass_rate_list=[sum(df[q])/examinee_num for q in q_num_list]

    #点双列相関係数は、項目jと得点yの標本共分散を項目の標本偏差と得点の標本偏差の積で割った値
    correlation_coefficient_list=[]
    for q in q_num_list:
        #項目jと得点yの不偏共分散から標本共分散を求める
        unbiased_covariance_jy=np.cov(df[q].tolist(), df["得点_y"].tolist())[0,1]#unbiased_covariance不偏分散
        biased_covariance_jy=unbiased_covariance_jy*(examinee_num-1)/examinee_num#biased_covariance標本分散

        #項目jの不偏分散から標本分散を求め、標準偏差を求める
        unbiased_covariance_j=np.cov(df[q].tolist(), df["得点_y"].tolist())[0,0]#unbiased_covariance不偏分散
        biased_covariance_j=unbiased_covariance_j*(examinee_num-1)/examinee_num
        standard_j=math.sqrt(biased_covariance_j)

        #得点yの不偏分散から標本分散を求め、標準偏差を求める
        unbiased_covariance_y=np.cov(df[q].tolist(), df["得点_y"].tolist())[1,1]#unbiased_covariance不偏分散
        biased_covariance_y=unbiased_covariance_y*(examinee_num-1)/examinee_num
        standard_y=math.sqrt(biased_covariance_y)

        correlation_coefficient_list.append(biased_covariance_jy/(standard_j*standard_y))

    PR_r_df=pd.DataFrame([pass_rate_list,correlation_coefficient_list],columns=q_num_list,index=['通過率', '点双列相関係数'])
    return PR_r_df
```

```python
start = time.time()
abc_para_list=[[initial_value_ab.values[0,q],initial_value_ab.values[1,q],0] for q in range(len(q_num_list))]
div=15*2
x_plot_list=np.linspace(-4,4,div).tolist()

series_list=[]

update_time=0
Amount_of_change_MT=np.matrix(np.ones((len(q_num_list),3)))

tf_zero=[0 for q in q_num_list]
tf_one=[1 for q in q_num_list]

while tf_zero != tf_one:
    print()

    
    G_im_list=G_im(abc_para_list)
    abc_para_list_before=np.matrix(abc_para_list)
    
    print(str(update_time)+"回目の更新")
    print(abc_para_list)

    N_r_list=N_r(G_im_list)
    N_list=[N_r_list[m][0] for m in range(div)]

    row=[]
    for q in range(len(q_num_list)):
        abc_para=abc_para_list[q]
        gosa_matrix=np.linalg.inv(matrix_I(abc_para))
        row.append(math.sqrt(gosa_matrix[0,0]))
        row.append(math.sqrt(gosa_matrix[1,1]))
    series_list.append(pd.Series(row, index=list(itertools.chain.from_iterable([[q+"a", q+"b"] for q in q_num_list]))))

    
    for q in range(len(q_num_list)):
        abc_para=abc_para_list[q]
        r_list=[N_r_list[m][q+1] for m in range(div)]

    #     print("項目"+str(q+1)+"のパラメータの初期値")
    #     print([abc_para[0],abc_para[1]])
    #     print("")

        Amount_of_change=np.matrix([[1],[1]])
        matrix_t=np.matrix([[abc_para[0]],[abc_para[1]]])
        

        while (abs(float(Amount_of_change[0])) > 0.00001 and abs(float(Amount_of_change[1])) > 0.00001) : #変化量が0.001大きい間は、繰り返し
            
            matrix_t_before = matrix_t
            matrix_t=matrix_t+np.linalg.inv(matrix_I(abc_para))*matrix_d(abc_para,r_list)
            matrix_t_after = matrix_t

            abc_para=[matrix_t[0,0], matrix_t[1,0], 0]

            Amount_of_change=matrix_t_before-matrix_t_after

        abc_para_list[q]=abc_para
    #     print("更新回数")
    #     print(update_time)
    #     print("")

    #     print("項目"+str(q+1)+"の推定値")

    #     print(abc_para_list[q])
    #     print("")
    #     print("-------------------------------------------------------------------------------")
    #     print("")

    print("")
    print(abc_para_list)

    abc_para_list_after=np.matrix(abc_para_list)
    
    #
    Amount_of_change_MT=abc_para_list_before-abc_para_list_after
    for q in range(len(q_num_list)):
        if abs(Amount_of_change_MT[q,0]) < 0.00001 and abs(Amount_of_change_MT[q,1]) < 0.00001:
            tf_zero[q]=1

    print(dict([(q_num_list[q], tf_zero[q]) for q in range(len(q_num_list))]))
    
    update_time=update_time+1
    
    print("")
    print("-------------------------------------------------------------------------------")
    print("")
    
gosa_df=pd.DataFrame(series_list)
estimate_df=pd.DataFrame([[abc_para_list[q][0] for q in range(len(q_num_list))],[abc_para_list[q][1] for q in range(len(q_num_list))]],columns=q_num_list,index=['推定値a', '推定値b',])
end = time.time()-start
print("処理時間：",end)

estimate_df
```
